```python
anscombe['x'].var()
```

```python
x_std=anscombe['x'].std()
x_mean=anscombe['x'].mean()
```

```python
y_std=anscombe['y'].std()
y_mean=anscombe['y'].mean()
```

```python
x_cv = x_std/x_mean
y_cv = y_std/y_mean
```

```python
print(f'The Standard Deviation of x: {x_std}')
print(f'The Mean of x: {x_mean}')
print(f'The standard deviation of y: {y_std}')
print(f'The Mean of y: {y_mean}')
print(f'The coefficient of variation of x: {x_cv}')
print(f'The coefficient of variation of y: {y_cv}')
```
